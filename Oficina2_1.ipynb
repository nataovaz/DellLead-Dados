```python
#Passo 1 - Importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
#Passo 1 - Importando JSON de dados
filmes_uri = 'https://d11s0xq1vqg039.cloudfront.net/videos_conteudo/visual/micro03/oficina/arquivos/dadosoficina03.txt'
filmes = pd.read_json(filmes_uri)
filmes.head()
```

```python
#Passo 2 - Crie uma visualização que mostre o orçamento de todos os filmes do conjunto de dados. Qual o filme com maior orçamento e com menor orçamento?
maior_orc = filmes[["Budget_M", "Film"]].sort_values(by='Budget_M', ascending=False)
maior_orc = maior_orc.head(10)

plt.title('Filmes com maior orçamento')
sns.barplot(data = maior_orc, y = 'Film', x='Budget_M')
```

```python
men_orc = filmes[["Budget_M", "Film"]].sort_values(by='Budget_M', ascending=True)
men_orc = men_orc.head(10)
plt.title('Filmes com menor orçamento')
sns.barplot(data = men_orc, y = 'Film', x='Budget_M')
```

```python
#Passo 3 - Crie uma visualização que mostre a bilheteria de todos os filmes do conjunto de dados. Qual o filme com maior bilheteria e com menor bilheteria?
maior_bil = filmes[["Worldwide_Gross_M", "Film"]].sort_values(by='Worldwide_Gross_M', ascending=False)
maior_bil = maior_bil.head(10)
plt.title('Filmes com maior bilheteria')
sns.barplot(data = maior_bil, y = 'Film', x='Worldwide_Gross_M')
```

```python

men_bil = filmes[["Worldwide_Gross_M", "Film"]].sort_values(by='Worldwide_Gross_M', ascending=True)
men_bil = men_bil.head(10)
plt.title('Filmes com maior bilheteria')
sns.barplot(data = men_bil, y = 'Film', x='Worldwide_Gross_M')
```

Passo 4 - Crie uma visualização que mostre a quantidade de filmes por gênero em cada ano. Em que anos foram feitos menos filmes de ação? E em que ano foram feitos menos filmes de animação?

```python
# display(ctg)
plt.figure(figsize=(15,8))
plt.title('Filmes com maior bilheteria')
sns.countplot(x='Year', hue='Genre', data=filmes)
```

Passo 5 - Crie uma visualização que mostre a média de orçamento de filmes por 
gênero em cada ano. Em que ano os filmes de ação tiveram maior média de orçamento? E em que ano os filmes de animação tiveram a maior média de orçamento?

```python
plt.figure(figsize=(10,5))
plt.title('Bilheteria média por gênero')
sns.lineplot(x='Year', y="Worldwide_Gross_M", hue='Genre', estimator = 'mean', data=filmes)
plt.show()
```
