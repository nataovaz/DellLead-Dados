```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
games = pd.read_csv('https://d11s0xq1vqg039.cloudfront.net/videos_conteudo/visual/micro04/oficina/arquivos/dadosoficina04.txt')
```

```python
display(games)

display(games.head())
```

```python
#Passo 2 - Crie uma visualização que mostre a proporção de vendas global entre gêneros de jogos da empresa Nintendo;
selecao = (games['Publisher'] == 'Nintendo')
selecao = games[selecao]
selecao = selecao.groupby(by='Genre').sum()

plt.figure(figsize=(15,8))
plt.title('Proporção de vendas global entre gêneros de jogos da empresa Nintendo')
plt.pie(x=selecao['Global_Sales'], labels = selecao.index, autopct='%1.1f%%',startangle=90, pctdistance=0.85)
plt.show()
```

```python
#Passo 3 - Crie uma visualização que mostre a proporção de quantidade de jogos por plataforma a partir do ano de 2015.
selecao = (games['Year'] >= 2015.0)
generos = games[selecao]
generos = generos.groupby(by='Platform').sum()

plt.figure(figsize=(15,8))
plt.title('Proporção de quantidade de jogos a partir de 2015')
plt.pie(x=generos['Global_Sales'], labels = generos.index, autopct='%1.1f%%',startangle=90, pctdistance=0.85)
plt.show()
```
